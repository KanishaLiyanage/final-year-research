## Tokenization

```python
vocabulary_size = 10000 # arbitrarily chosen
embed_size = 32 # arbitrarily chosen
```

**Creating & fitting Tokenizer**

```python
sinhala_tokens = Tokenizer(num_words = vocabulary_size)
```

```python
sinhala_tokens.fit_on_texts(X_train.values)
```

**Tokenizing texts**

```python
X_train = sinhala_tokens.texts_to_sequences(X_train)
X_test = sinhala_tokens.texts_to_sequences(X_test)
```

**Padding texts**

```python
X_train = pad_sequences(X_train, maxlen = maxlen)
X_test = pad_sequences(X_test, maxlen = maxlen)
```

**Saving the tokenizer**

```python
# Convert the tokenizer to a JSON string
tokenizer_json = sinhala_tokens.to_json()

# Save the JSON string to a file
with open('7_classes_tokenizer.json', 'w', encoding='utf-8') as json_file:
    json.dump(tokenizer_json, json_file, ensure_ascii=False)
```

## Bidirectional LSTM Model

```python
model = Sequential()

# Input Layer
model.add(Embedding(vocabulary_size, embed_size, input_length = maxlen))

model.add(Bidirectional(LSTM(128, return_sequences = True))) 
model.add(Bidirectional(LSTM(64, return_sequences = True)))

model.add(GlobalMaxPooling1D()) 

model.add(Dense(1024))
model.add(Dropout(0.25)) 
model.add(Dense(512))
model.add(Dropout(0.25))
model.add(Dense(256))
model.add(Dropout(0.25))
model.add(Dense(128))
model.add(Dropout(0.25))
model.add(Dense(64))
model.add(Dropout(0.25))

# Output Layer
model.add(Dense(classes, activation = 'softmax'))
```

```python
model.summary()
```

```python
plot_model(model, to_file = '7_classes_bilstm_model.png', show_shapes=True, show_layer_names=True)
```

```python
callbacks = [
    
             EarlyStopping(
                           monitor = 'val_accuracy',
                           min_delta = 1e-4,
                           patience = 4,
                           verbose = 1
                           ),

             ModelCheckpoint(
                             filepath = '7_classes_bilstm_best_model.h5',
                             monitor = 'val_accuracy', 
                             mode = 'max', 
                             save_best_only = True,
                             save_weights_only = False, # Set to False to save the entire model
                             verbose = 1
                             )
    
            ]
```

```python
model.compile(
              loss = 'sparse_categorical_crossentropy',
              optimizer = 'adam', 
              metrics = ['accuracy']
             )
```

```python
hist = model.fit(
                 X_train,
                 y_train,
                 batch_size = batch_size,
                 validation_data = validation_data,
                 epochs = epochs,
                 callbacks = callbacks
                )
```

**Accuracy and loss plots**

```python
fig , ax = plt.subplots(1,2)
train_acc = hist.history['accuracy']
train_loss = hist.history['loss']
fig.set_size_inches(12,4)

ax[0].plot(hist.history['accuracy'])
ax[0].plot(hist.history['val_accuracy'])
ax[0].set_title('Training Accuracy vs Validation Accuracy')
ax[0].set_ylabel('Accuracy')
ax[0].set_xlabel('Epoch')
ax[0].legend(['Train', 'Validation'], loc = 'upper left')

ax[1].plot(hist.history['loss'])
ax[1].plot(hist.history['val_loss'])
ax[1].set_title('Training Loss vs Validation Loss')
ax[1].set_ylabel('Loss')
ax[1].set_xlabel('Epoch')
ax[1].legend(['Train', 'Validation'], loc = 'upper left')

plt.show()
```

```python
# Input layer
input_layer = Input(shape=(maxlen,))

# Shared embedding layer
embedding = Embedding(vocabulary_size, embed_size, input_length=maxlen)(input_layer)
spatial_dropout = SpatialDropout1D(0.1)(embedding)

# TCN branch
tcn = TCN(128, return_sequences=True, dilations=[1, 2, 4, 8], name='tcn1')(spatial_dropout)
tcn = TCN(64, return_sequences=True, dilations=[1, 2, 4, 8], name='tcn2')(tcn)
tcn_pool = GlobalMaxPooling1D()(tcn)

# BiLSTM branch
lstm = Bidirectional(LSTM(128, return_sequences=True))(spatial_dropout)
lstm = Bidirectional(LSTM(64, return_sequences=True))(lstm)
lstm_pool = GlobalMaxPooling1D()(lstm)

# Concatenate the outputs of TCN and BiLSTM branches
merged = concatenate([tcn_pool, lstm_pool])

# Dense layers
dense = Dense(1024, activation='relu')(merged)
dense = Dropout(0.25)(dense)
dense = Dense(512, activation='relu')(dense)
dense = Dropout(0.25)(dense)
dense = Dense(256, activation='relu')(dense)
dense = Dropout(0.25)(dense)
dense = Dense(128, activation='relu')(dense)
dense = Dropout(0.25)(dense)
dense = Dense(64, activation='relu')(dense)
dense = Dropout(0.25)(dense)

# Output layer
output = Dense(classes, activation='softmax')(dense)

# Compile model
model = Model(inputs=input_layer, outputs=output)
```
