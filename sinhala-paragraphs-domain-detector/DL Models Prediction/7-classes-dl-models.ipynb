```python
from keras.utils.vis_utils import plot_model
```

## Importing Trained Models

```python
# Bidirectional LSTM
bilstm_model = load_model('../Trained Models for 07 Classes/Bidirectional LSTM/7_classes_bilstm_best_model.h5')

# Bidirectional GRU
bigru_model = load_model('../Trained Models for 07 Classes/Bidirectional GRU/7_classes_bigru_best_model.h5')

# Ensemble TCN-BiLSTM
ensemble_tcnbilstm_model = load_model('../Trained Models for 07 Classes/Ensemble TCN-BiLSTM/7_classes_ensemble_tcnbilstm_best_model.h5', custom_objects={'TCN': TCN})

# Ensemble TCN-BiGRU
ensemble_tcnbigru_model = load_model('../Trained Models for 07 Classes/Ensemble TCN-BiGRU/7_classes_ensemble_tcnbigru_best_model.h5', custom_objects={'TCN': TCN})

# TCN
tcn_model = load_model('../Trained Models for 07 Classes/TCN/7_classes_tcn_best_model.h5', custom_objects={'TCN': TCN})
```

## Bidirectional LSTM Model Prediction

```python
plot_model(bilstm_model, to_file='bilstm_model.png', show_shapes=True, show_layer_names=True)
```

## Bidirectional GRU Model Prediction

```python
plot_model(bigru_model, to_file='bigru_model.png', show_shapes=True, show_layer_names=True)
```

## TCN Model Prediction

```python
plot_model(tcn_model, to_file='tcn_model.png', show_shapes=True, show_layer_names=True)
```

## Ensemble TCN-BiLSTM Model Prediction

```python
plot_model(ensemble_tcnbilstm_model, to_file='ensemble_tcnbilstm_model.png', show_shapes=True, show_layer_names=True)
```

## Ensemble TCN-BiGRU Model Prediction

```python
plot_model(ensemble_tcnbigru_model, to_file='ensemble_tcnbigru_model.png', show_shapes=True, show_layer_names=True)
```
